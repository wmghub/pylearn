# 3-1. 辞書 (dictionary)

**辞書**は、**キー** (**key**) と**値** (**value**) を対応づけるデータです。
キーとして、文字列・数値・タプルなどの変更不可能なデータを使うことができます。<br>
キーは名前どおりキーですので、キーに変更可能なデータであるリスト・辞書を使うことはできません。<br>
値としては、あらゆる種類のデータを指定できます。<br>
たとえば、文字列 `'apple'` をキーとし値として数値 `3` を、`'pen'` をキーとして `5` を対応付けた辞書は、<br>
次のように作成します。

```python
ppap = { (1,2,3) : 'pineapple', 'apple' : 3, 'pen' : 5}
print("dictionary contents is", ppap)
print("dictionay type is", type(ppap))
print("key: apple value:", ppap['apple']) # 辞書[キー値]で紐づく値（バリュー）を取得できます

# キーに対する値を変更したり、新たなキーと値を登録するには代入文を用います。
ppap['apple'] = 10
ppap['pinapple'] = 7
print("dictionary contents is", ppap)
```

```python
print("key: orange value:", ppap['orange']) # 辞書に登録されていないキーを指定すると、エラーになります。
```

キーから値は取り出せますが、値からキーを直接取り出すことはできません。<br>
また、リストのようにインデックスを指定して値を取得することはできません。

```python
ppap[1]
```

キーが辞書に登録されているかどうかは、演算子 **`in`** を用いて調べることができます。

```python
ppap = {'apple': 3, 'pen': 5}
print("ppap contains apple:", 'apple' in ppap)
print("ppap contains banana:", 'banana' in ppap)
```

組み込み関数 **`len`** によって、辞書に登録されている要素、キーと値のペア、の数が得られます。

```python
ppap = {'apple': 3, 'pen': 5}
print("ppap contains key and value pair(s): ", len(ppap))
del ppap['pen']
print("ppap:", ppap)
```

空の辞書も作れます。

```python
empty_d = {}
empty_d
```

## 辞書のメソッド

### キーを指定して値を得るメソッド

**`get`** メソッドは引数のキーが辞書に含まれていない場合に `None` を返します。<Br>
`get` を利用することで、エラーを回避できます。<br>
キーを括弧、`[...]`、で指定する方法では、<br>
辞書にキーが存在しないとエラーとなりプログラムの実行が停止します。

```python
ppap = {'apple' : 3, 'pen' : 5}
print('キーappleに対応する値 = ', ppap.get('apple'))
print('キーorangeに対応する値 = ', ppap.get('orange'))
print('キーorangeに対応する値（エラー） = ', ppap['orange'])
```

また、`get` に2番目の引数を与えると、その引数の値を「指定したキーが含まれていない場合」に `get` が返す値とすることができます。

```python
ppap = {'apple' : 3, 'pen' : 5}
print('キーappleに対応する値 = ', ppap.get('apple', -1))
print('キーorangeに対応する値 = ', ppap.get('orange', -1))
```

### キーの一覧を得る
**`keys`** メソッドはキーの一覧を返します。<br>

次のように、**`keys`** メソッドが返した結果に関数 `list` を適用すると、
通常のリストになります。<br>
<br>
次のプログラムでキーと値のペアを定義してみてください。

```python
ppap = {'apple' : 3, 'pen' : 5}
list(ppap.keys()) # list(`要素`)でリストを作成する
```

### 値の一覧を得る
**`values`** メソッドはキーに対応する全ての値の一覧を返します。<br>

次のプログラムでキーと値のペアを定義してみてください。

```python
ppap = {'apple' : 3, 'pen' : 5, 'orange': 200}
list(ppap.values())
```

### キーと値の一覧を得る
**`items`** メソッドはキーとそれに対応する値をタプルにした一覧を返します。 <br>
タプルのリストを作成できます。

```python
ppap = {'apple' : 3, 'pen' : 5}
list(ppap.items())
```

## 辞書とリスト

次のように値としてリストを使用する辞書を作成できます。<br>
リストの要素を参照するには数字インデックスをさらに指定します。

```python
numbers = {'dozens': [10, 20, 40], 'hundreds': [100, 101, 120, 140]}
print(numbers['dozens'])
print(numbers['hundreds'][2])
```

辞書を要素とするリストを作成することもできます。

```python
ppap = {'apple': 3, 'pen': 5}
pets = {'cat': 3, 'dog': 3, 'elephant': 8}
ld = [ppap, pets]
print(ld[1])
print(ld[1]['dog'])
```
