## 練習
リスト `ln` を引数として取り、`ln` の偶数番目のインデックスの値を削除したリストを返す関数 `remove_evenindex` を作成してください（ただし、0 は偶数として扱うものとします）。

ヒント：スライスを使います。

以下のセルの `...` のところを書き換えて `remove_evenindex(ln)` を作成してください。

````python
def remove_evenindex(ln):
#    print(ln[1::2])
    return ln[1: :2] # リストの1番目からインデックスを２要素おきに抜き出します
````

上のセルで解答を作成した後、以下のセルを実行し、実行結果が全て `True` になることを確認してください。

````python
print(remove_evenindex(['a', 'b', 'c', 'd', 'e', 'f', 'g']) == ['b', 'd', 'f'] )
print(remove_evenindex([1, 2, 3, 4, 5]) == [2, 4])
````

### 練習

英語の１文からなる文字列 `str_engsentence` が引数として与えられたとき、`str_engsentence` 中に含まれる単語数を返す関数 `count_words` を作成してください。ただし、文はピリオドで終了し単語は空白で区切られるものとします。

ヒント：文字列に対する `split` メソッドを使います。


````python
def count_words(str_engsentence):
  tokens = str_engsentence.split(' ')
  return len(tokens)
````

### リストと演算子

演算子 **`+`** によってリストの連結、**`*`** によって連結における繰り返し回数を指定することができます。

````python
numbers = [0, 10, 20, 30, 40, 50]
mix = numbers + ['a', 'b', 'c']
print("mix is", mix)
numbers *= 3 # `numbers = numbers * 3` と同じ
print("3 times", numbers) # 連結を3回繰り返す
````

演算子 **`in`** は、左辺の要素がリストに含まれれば `True` を、それ以外では `False` を返します。

````python
numbers = [0, 10, 20, 30, 40, 50]
10 in numbers
````

リストに対する `in` 演算子は、論理演算 `or` を簡潔に記述するのに用いることもできます。 たとえば、

---
```Python
a1 == 1 or a1 == 3 or a1 == 7:
```
---

は

---
```Python
a1 in [1, 3, 7]:
```
---

と同じ結果を得られます。 `or` の数が多くなる場合は、`in` を用いた方がより読みやすいプログラムを書くことができます。

**`not in`** 演算子は、`in` 演算子の逆を意味します。

````python
10 not in numbers
````

````python
11 not in numbers
````

### 並べ替え（**`sort`** メソッド）

`sort` メソッドはリスト内の要素を並べ替えます。
引数に何も指定しなければ昇順でとなります。

````python
numbers = [30, 50, 10, 20, 40, 60]
numbers.sort()
characters = ['e', 'd', 'a', 'c', 'f', 'b']
characters.sort()
print("numbers sorted as", numbers)
print("characters sorted as", characters)
print("numbers sorted by descendents as", numbers.sort(reverse = True)) # reverse = True オプションを指定すれば、要素を降順に並べ替えることもできます
print("characters sorted by descendents as", characters.sort(reverse = True)) # reverse = True オプションを指定すれば、要素を降順に並べ替えることもできます
````

### 並べ替え（**`sorted`** 組み込み関数）

関数 `sorted` ではリストを引数に取って、そのリスト内の要素を昇順に並べ替えた結果をリストとして返します。

---
```Python
sorted(リスト)
```
---

````python
numbers = [30, 50, 10, 20, 40, 60]
characters = ['e', 'd', 'a', 'c', 'f', 'b']
print("numbers are sorted as", sorted(numbers))
print("characters are sorted as", sorted(characters)) # 文字列の比較は辞書順
print("numbers", numbers)
print("characters", characters)
print()
print("numbers are sorted by descendents as", sorted(numbers, reverse=True))
print("characters are sorted by descendents as", sorted(characters, reverse=True))
````

多重リストをソートするとどのような結果が得られるか確かめてみてください。

````python
lns = [[20, 5], [10, 30], [40, 20], [30, 10]]
lns.sort()
lns
````

## 破壊的（インプレース）な操作と非破壊的な生成

` sort` メソッドでは元のリストが変更されています。<br>
一方、`sorted` 関数では元のリストはそのままになっています。

````python
numbers = [30, 50, 10, 20, 40, 60]
numbers.sort()
print('sortメソッドの実行後の元のリスト:', numbers)
numbers = [30, 50, 10, 20, 40, 60]
sorted(numbers)
print('sorted関数の実行後の元のリスト:', numbers)
````

`sort` メソッドは元のリストを変更してしまいます。
これを**破壊的**あるいは**インプレース** (**in-place**) な操作と言います。<br>
一方、`sorted` 関数は新しいリストを生成し元のリストを破壊しません、これを**非破壊的**であると言います。

## タプル (tuple)

**タプル**は、リストと同じようにデータの並びです。<br>
リストと違ってタプルは一度設定した要素を変更できません。<br>
タプルは変更不可能なデータです。<br>
作成した後で要素を変更する必要がない場合は、<br>
タプルの方が実装の効率がよいので、リストよりもタプルを使うべきです。<br>

たとえば、関数が複数の値をリストにして返し、呼び出し側がすぐにリストをばらばらにして値を取り出すような場合は、
リストよりもタプルを使うべきです。<br>
また、平面上の点を表そうとするとき、x座標とy座標を別々に変化させる必要がなければ、
`(3, 5)` のようなタプルを使うのが自然です。<br>
タプルを作成するには要素を丸括弧 `(...)` で囲みます。

````python
x = 3
y = 5
point = (x, y)
print(point)
print(type(point))
numbers3 = 1, 2, 3 # 丸かっこなしでもタプルは作成できます
print(numbers3)
print("only one element", (1)) # 要素が１つだけのタプル
print("empty tuple", ()) # 空のタプルß
````

リストや文字列と同様に、インデックスや組み込み関数を使った操作が可能です。

````python
numbers3 = (1, 2, 3)
print("index 1 ==", numbers3[1]) # インデックスの指定による値の取得
print("length ==", len(numbers3)) # lenはタプルを構成する要素の数
print("slice 1 to 3 ==", numbers3[1:3]) # スライス
````

組み込み関数 **`list`** を使って、タプルをリストに変換できます。
（`list` はリストのデータ型でもあります。）<br>
組み込み関数 **`tuple`** を使って、逆にリストをタプルに変換できます。
（`tuple` はタプルのデータ型でもあります。）

````python
numbers3 = (1, 2, 3)
print("tuple conver to list", list(numbers3)) # list に変換
numbers2 = [1,2]
print("list convert to tuple", tuple(numbers2)) # tuple に変換
````

## リストやタプルの比較演算

## for文による繰り返しとリスト・タプル

リストの要素のそれぞれを操作したい場合は**for文**を用います。
リスト `ls` の要素全てに何らかの操作を`実行` するにはループを使います。

---
```Python
for value in list:
    実行文
```
---

`for` で始まる行の `in` の後に処理対象となるリスト `list` が、`in` の前に変数 `value` が書かれます。<br>
`list` の最初の要素、すなわち `list[0]` が `value` に代入され `実行文` が処理されます。<br>
`実行文` の処理が終われば、`list` の次の要素が `value` に代入され、処理が繰り返されます。<br>
このようにして、`list` の要素に対する処理が `len(list)` 回繰り返されると、for文の処理が終了します。<br>　
ループでの `in` は、先に説明したリスト要素の有無の検査 `in` とは違います。<br>
また、if文と同様、 `実行文` の前にはスペースによるインデント（字下げ）が必要です。

次の例をみてみましょう。<br>
3つの要素を持つリスト `ls` から1つずつ要素を取り出し、変数 `value` に代入しています。
`実行文` では `value` を用いて取り出した要素を参照しています。

````python
list = [0,1,2]

for value in list:
    print('For loop:', value)
````

`squares1` には最初に空リストが代入されます。
そして、`numbers` の各要素の自乗がこのリストに次々と追加されます。

````python
numbers = [0,1,2,3,4,5]
square1 = []
for x in numbers:
    square1.append( x ** 2 )

print(square1)
````

## 練習
整数の要素からなるリスト `ln` を引数として取り、`ln` の要素の総和を返す関数 `sum_list` を作成してください。


````python
def sum_list(ln):
    # 合計値を格納する変数
    sum = 0

    for elm in ln: # リストlnの中の要素を elm に格納する
        #  要素を合計する
        sum += elm

    return sum # 合計を返す
````

上のセルで解答を作成した後、以下のセルを実行し、実行結果が全て True になることを確認してください。

````python
print(sum_list([10, 20, 30]) == 60)
print(sum_list([-1, 2, -3, 4, -5]) == -3)
````

## リスト初期化の短縮記法

先に、リストの要素の自乗から成るリストを求める例を説明しました。

````python
numbers = [0,1,2,3,4,5]

squares1 = []
for x in numbers:
    squares1.append( x ** 2 )

print(squares1)
````

````python
squares2 = [  x ** 2  for  x  in  numbers  ]
squares2
````

```Python
[x を含む式 for x in リストを返す式]
```
という式は、"リストを返す式" のリストの各要素を `x` に代入して "`x` を含む式" を計算し、<br>
その結果をリストにして返します。
