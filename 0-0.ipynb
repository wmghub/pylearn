# CLONEする
## github上のレポジトリをいちばん最初にダウンロードしてきます

自分の GoogleDrive に、指定した場所（下のセルでは、マイドライブ ＞ Colab Notebooks ＞ pylearn を指定しています。）に、ダウンロードします。<br>
git では、最初にダウンロードすることを clone （クローン）といいます。<br>
以降、自分の GoogleDrive に自分で指定したフォルダと github.com のレポジトリの間で<br>
どちらかが更新された際に生じる差分をダウンロード（pull）したり、<br>
アップロード（push）したりして、<br>
同期をとること（差分をなくして同じ状態にすること）が簡単にできるようになります。<br>

````python
from google.colab import drive
googledrive = '/content/drive'
drive.mount(googledrive)

%cd /content/drive/MyDrive/Colab\ Notebooks

!git clone https://github.com/wmghub/pylearn.git
````

# *PULLする*
---
## github上のレポジトリとローカルのレポジトリとの差分をダウンロードします

自分の GoogleDrive に指定した場所にすでにクローンしたレポジトリと<br>
クローンしてきたのと同じ github のレポジトリに差分があれば、ダウンロードしてきます。<br>
<br>

---
<hr>
このコードセルには、ファイルの更新を取り消すコマンドが含まれています。<br>

```Shell

```


````python
from google.colab import drive # GoogleDrive にアクセスするためのライブラリをインポート
googledrive = '/content/drive'   # 自分の GoogleDrive のマウントポイント
drive.mount(googledrive)

# 自分のGoogleDrive のローカルレポジトリのフォルダ位置
%cd /content/drive/MyDrive/Colab\ Notebooks/pylearn

# github のリモートレポジトリと比較したローカルにあるレポジトリの状態を見ます
!git reset --hard
!git pull
````

# レポジトリを作成する
## レポジトリを初期登録します

github でアカウントを作成して、user.emailとuser.nameを設定します。<br>
githubで作成したアカウントの Private Access Tokenを取得してください。<br>

````python
from google.colab import drive
googledrive = '/content/drive'
drive.mount(googledrive)
mydrive = '/content/drive/MyDrive'           # GoogleDrive の "マイドライブ"
colab_dir = mydrive + '/Colab Notebooks'  # Colab Notebooks のフォルダ（＝ディレクトリ名）
pylearn_dir = colab_dir + '/pylearn'            # githubに登録するレポジトリ名（＝フォルダ名）

%cd /content/drive/My Drive/Colab Notebooks/pylearn

!git config --global user.email "（※ここにgithubに登録する電子メールアドレスを入力）" # "warashina.makoto+github@gmail.com"
!git config --global user.name "（※ここにgithubに登録した名前を入力）" # "wmghub"

# github で取得した Personal Access Token をクォーテーションで囲わないでそのまま入力 XXXX... を置き換えます
# PASTE TOKEN ↓ここにパーソナルアクセストークンをペースト（貼り付け）してください
%env TOKEN=XXXXXXXXXXXXXXXXXXXXXXXXXXXXXX

# wmgghubは、githubユーザー名
# pylearn.git の pylearn の部分は githubのレポジトリ名
!git remote set-url origin https://wmghub:$$TOKEN@github.com/wmghub/pylearn.git
!git init
!git add README.md
!git commit -m "first commit"
!git branch -M main
!git remote add origin https://$$TOKEN@github.com/wmghub/pylearn.git
!git push -u origin main
````

# PUSHする
## 更新した内容を github に登録します

自分の GoogleDrive にあるレポジトリの内容を github に登録します。<br>
githubの自分のレポジトリか（Personal Access Tokenで個人を同定します）、<br>
許可されたレポジトリにのみ push することが可能です。<br>

````python
from google.colab import drive
googledrive = '/content/drive'
drive.mount(googledrive)

%cd /content/drive/MyDrive/Colab\ Notebooks/pylearn

!git config --global user.email "warashina.makoto+github@gmail.com"
!git config --global user.name "wmghub"

# github で取得した Personal Access Token をクォーテーションで囲わないでそのまま入力 XXXX... を置き換えます
# PASTE TOKEN ↓ここにパーソナルアクセストークンをペースト（貼り付け）してください
%env TOKEN=XXXXXXXXXXXXXXXXXXXXXXXXXXXXXX

!git remote set-url origin https://wmghub:$$TOKEN@github.com/wmghub/pylearn.git

# 変更を登録するファイルを追加する
!git add 0-0,jpynb

# ファイルの変更を登録する際のコメント
!git commit -m "feat: 変更"

# ファイルを登録する
!git push
````
