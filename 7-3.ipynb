## 線形回帰モデルを作成する
- 重回帰分析

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

values = worksheet.get_all_values()
pd.options.display.float_format = '{:}'.format

df = pd.DataFrame.from_records(data=values[1:])
df = df.astype({ 0: str, 1: int, 2: str, 3: str, 4: int, 5: int, 6: int, 7: int, 8: int, 9: int, 10: int, 11: int, 12: float })

adachi_df = df[[ 1, 4, 5, 6, 11, 12 ]] # 種別、最寄り所要時間、築年数、建物自体の階数、間取り、平米数
X = adachi_df
y = df[8]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression(fit_intercept=True, normalize=False, copy_X=True, n_jobs=None)
lr.fit(X_train, y_train)

print("Score:{}".format(lr.score(X_test, y_test)))

print("prediction", format(  lr.predict(  [[3, 13, 24, 3, 16, 68.6]]  )  ))

print("Intercept:{}".format(lr.intercept_))
print("Coefficients:{}".format(lr.coef_))
```
