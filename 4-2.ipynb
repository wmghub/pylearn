## 4-2. イテレータ
イテレータについて簡単に説明します。

## `next`

ファイルオブジェクトは、**イテレータ**と呼ばれるオブジェクトの一種です。英単語の`iterate`は繰り返すという意味です。<br>
イテレータオブジェクトは、要素を1つずつ取り出す処理が可能です。<br>
**`next`** という関数でその処理を1回分行います。<br>

変数 `f` にファイルオブジェクトが入っているとすると、<br>
`next(f)` は、ファイルから新たに1行を読んで文字列として返します。

```python
with open(pylearn_dir + '/sample.txt', 'r') as f:
    for line in f:                                  # StopIterationエラーを検知してループを抜ける
        print(line, end='')
```

`ファイルオブジェクト.readline()` と `next(ファイルオブジェクト)` は、ファイルから1行を読んで文字列として返します。<br>
`readline()` と `next()` では、ファイルの終わりに来たときの挙動が異なります。<br>
`readline()` は `''` という空文字列を返します。<br>
`next()` は **`StopIteration`** というエラーを送出します。<br>
`next(f)` が `StopIteration` を送出したらforループから抜け出します。<br>

このように、関数 `next` が適用できて、<br>
`next` が何らかの値を返すか、`StopIteration` を発するようなオブジェクトを、イテレータと呼びます。

## 練習
文字列 `name` をファイル名とするファイルの最後の行を文字列として返す関数 `last_line(name)` を、
ファイルオブジェクトに対するfor文を用いて定義してください。

```python
# 関数の定義
def last_line(name):
    with open(pylearn_dir + '/' + name) as f:
        for line in f:
            current_line = line
    return current_line

# 関数の検証
print(last_line('sample.txt')=="Excepteur sint occaecat cupidatat non proident, sunt in culpa qui officia deserunt mollit anim id est laborum.\n")
```

## `iter`

組み込み関数 **`iter`** でイテレータを作成することができます。<br>


```python
it = iter([1,2,3])
print('=== 最初のイテレーション')
for elm in it:
    print(elm)
print('=== ２回目のイテレーション') # ファイルの場合と同様に、もう一回for文を回しても何も出力されません
for elm in it:
    print(elm)

print("=== 最初の forループ")
li = [1,2,3]
for elm in li:
    print(elm)
print("=== ２回目の forループ")
for elm in li:
    print(elm)
```

## 練習

リストをもらって、そのイテレータを作り、最初の要素だけ取り出した後、
そのイテレータを返す関数 `but_first(ls)` を定義してください。

```python
# 関数の定義
def but_first(ls):
    it = iter(ls)
    for i in range(0, 1):
        next(it)
    return it

# 関数の検証
it = but_first([0,2,4,6,8])
print(type(it) == type(iter([]))) # タイプはリストイテレータ
print(list(it) == [2,4,6,8])
```

## イテラブル

**イテラブル** オブジェクトは、関数 `iter` が適用できます。<br>
イテラブルは、for文の `in` の後に指定することができます。<br>

リストはイテラブルですが、イテレータではありません。

```python
ln = [1,2,3]
for x in ln:
    print(x)
for x in ln:
    print(x)
```

上の例では、2つのfor文ごとに、`ln` に `iter` が適用されて、別々のイテレータが作られます。

関数 **`range`** が返すオブジェクトもイテラブルですが、イテレータではありません。

```python
r = range(3)
for x in r:
    print(x)
for x in r:
    print(x)
```

## イテレータを返す `enumerate`

**`enumerate()`** は、イテレータを返す組み込み関数です。

```python
with open(pylearn_dir + '/sample.txt', 'r') as f:
    for i, s in enumerate(f):
        print(i, '行目:')
        print(s)
```
