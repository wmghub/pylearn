## ファイルのオープン
**ファイル**を読み込んだり、書き込んだりするには、<br>
まず、**`open()`** という関数によってファイルを**オープン**する必要があります。<br>
<br>
次のセルで変数 `f` には、**ファイルオブジェクト**が代入されます。このオブジェクトを通じてファイルを読み書きします。<br>
`'sample.txt'` はファイル名で、そのファイルの絶対パス名か、このノートブックからの相対パス名を指定します。<br>
`'r'` はファイルをどのモードで開くかを指しており、`'r'`は**読み込みモード**を意味します。<br>
読み込みモードでオープしたファイルに書き込みすることはできません。<br>
<br>
モードには次のような種類があります。<br>

記号| モード
---|:---
r | 読み込み
w | 書き込み
a | 追記
+ | 読み書き両方を指定したい場合に使用

```python
f = open(pylearn_dir + '/sample.txt', 'r')
```

```python
f.close()
```

## 行の読み込み
ファイルオブジェクトの、**`readline()`** という関数を呼び出すと、ファイルから新たに1行を読んで文字列として返します。<br>
読み込んでくる１行は、ファイルの先頭もしくは改行文字の次の文字から、ファイルの終わり、または改行文字（'\n'）までの文字列です。<br>
<br>
ファイルの終わりに来たとき、`readline()` は `''` という空文字列を返します。<br>
次のようにして `readline()` を使ってファイルを行単位で読んでみましょう。<br>
`readline()` 関数はファイルオブジェクトを消費します。<br>
改めてファイルの先頭から読み出したいときには、再度オープンする必要があります。

```python
f = open(pylearn_dir + '/sample.txt', 'r')
i = 0

while True:
    line = f.readline()
    i += 1
    if line != '':
        print('line[' + str(i) + ']', line)
    else:
        f.close()
        break
```

## 練習
文字列 `name` をファイル名とするファイルの最後の行を文字列として返す関数 `last_line(name)` を定義してください。

```python
# 関数の定義
def last_line(name):
    f = open(pylearn_dir + '/' + name)
    while True:
        line = f.readline()
        if line != '':                   # ファイルの終端でない場合
            previous_line = line # 現在行の内容を、変数`前行`（previous_line）に代入
        else:                           # ファイルの終端の場合
            f.close()                  # ファイルをクローズ
            break                     # ループを抜ける
    return previous_line        # ループを抜けた時は現在行は空文字（ '' ）のため前行を返す

# 関数の検証
# lineline() では、行末の '\n' まで返されることに注意
print(last_line('sample.txt')=="Excepteur sint occaecat cupidatat non proident, sunt in culpa qui officia deserunt mollit anim id est laborum.\n")
```

## ファイルへの書き込み

ファイルへの書き込みには、`print` 関数を使えます。<br>
`file` 引数に書き込み先のファイルオブジェクトをキーワード引数で指定します。

```python
with open(pylearn_dir + '/print-test.txt', 'w') as f:
    print('hello\nworld', file=f)
```

ファイルをオープンする際にモードを**書き込みモード**の `'w'` に指定しています。同名のファイルがあれば上書き、なければ新規作成します。<br>
モードが`'a'` の場合は、ファイルがあれば中身を追記し、なければ新規作成します。<br>
<br>
`print` 関数は、デフォルトで文字列の末尾に改行文字を加えて印字します。<br>
末尾に加える文字は、`end` 引数で指定できます。

```python
with open(pylearn_dir + '/print-test.txt', 'a') as f:
    print('hello', 'world\n', end='', file=f) # 改行文字を加えない
```

また、複数の印字対象を渡すと、デフォルトで、空白文字で区切って印字します。
この区切り文字は、`sep` 引数で指定できます。

```python
with open(pylearn_dir + '/print-test.txt', 'a') as f:
    print('hello', 'world', sep='\n', file=f) # 'hello, world'が印字される
```

この他にも、ファイルオブジェクトには、１文字ずつ書き込む`write()` 関数と１文字ずつ読み出す `read()` 関数があります。<br>
次のコードは、`sample.txt` を `sample.txt.bak` にコピーします。<br>

```python
with open(pylearn_dir + '/sample.txt') as src, open(PYTHON_DIR + '/sample.backup.txt', 'w') as dst:
    dst.write(src.read())
```

## 練習

2つのファイル名 `infile`, `outfile` を引数として、`infile` の半角英文字を全て大文字にした結果を `outfile` に書き込む `file_upper(infile, outfile)` という関数を作成してください。   

なお、半角英文字の小文字を大文字に変換するには `upper()` というメソッドが使えます。
たとえば `line` という名前の変数に半角文字列が入っている場合、`line.upper()` とすれば小文字に変換した文字列を返します。

```python
# 関数の定義
def file_upper(infile,outfile):
    with open(pylearn_dir + '/' + infile, 'r') as in_fileobj, open(PYTHON_DIR + '/' + outfile, 'w') as out_fileobj:
        while True:
            in_line = in_fileobj.readline()
            if in_line != '':
                out_line = in_line.upper()
                print(out_line, end='', file=out_fileobj) # print関数は、行末にデフォルトで'\n'を追加するため、空文字を追加するように指定
            else:
                break

# 検証のためにファイルを作成する
with open(pylearn_dir + '/print-test.txt', 'w') as f:
    print('hello', 'world', file=f)

# 関数の実行
file_upper('print-test.txt', 'print-test-upper.txt')

# 関数の検証
with open(pylearn_dir + '/print-test-upper.txt', 'r') as f:
    print(f.read() == 'HELLO WORLD\n')
```

## ファイルの読み書きにおける文字コード指定

`open` でファイルを開くと、テキストモードで開きます（テキストモード以外にバイナリモードもあります）。<br>
<br>
テキストモードではデフォルトの**文字コード**でファイルを開きます。<br>
文字コードがファイルを書き込む際に指定したものと異なる場合、エラーが出たり文字化けします。<br>
<br>
デフォルトの文字コードは、WindowsはShift_JIS、macOSやLinuxはUTF-8になっていることが多いです。<br>
UTF-8で文字を記録されたファイルをWindowsで、ただ `open('utf-8.txt', 'w')` のように文字コードを指定せずに開くとエラーが出ます。<br>
同じく、Shift_JISで文字を記録されたファイルをmacOSで `open('shift_jis.txt', 'w')` として開くとエラーが出ます。<br>
<br>
`sample.txt` では半角英数字しか使われておらず、半角英数字に関しては、Shift_JISもUTF-8も共通のASCIIコードんで文字がエンコードされているためです。

```python
# Google Colaboratory では shift_jis でエンコードされたテキストファイルを開こうとするとエラー
# macOS ならこちらでエラー
with open(pylearn_dir + '/shift_jis.txt', 'r') as f:
    print(f.read())

# Windowsならこちらでエラー
with open(pylearn_dir + '/utf-8.txt', 'r') as f:
    print(f.read())
```

日本語のテキストファイルを開く時は、ファイル内の文字コードと同じ文字コードを指定してください。<br>

文字コード指定は、`open` のキーワード引数で
`encoding='utf-8'`（文字コードにUTF-8を指定する場合）のように指定することができます。 <br>

日本語の文字コードには `UTF-8`, `Shift_JIS`, `EUC-JP` などがあります。<br>
PythonではOSの種類に限らず、UTF-8という文字コードがよく使われます。<br>
Google Colaboratoryでも UTF-8が使われています。

```python
# 文字コードを指定しないとmacOSならこちらでエラー
# Google Colaboratory でもエラー
with open(pylearn_dir + '/shift_jis.txt', 'r', encoding='shift_jis') as f:
    print(f.read())

# 文字コードを指定しないとWindowsならこちらでエラー
with open(pylearn_dir + '/utf-8.txt', 'r', encoding='utf-8') as f:
    print(f.read())

# 文字コードを指定してファイルに書き込む場合
with open(pylearn_dir + '/text.txt', 'w', encoding='utf-8') as f:
    f.write('かきくけこ')
with open(pylearn_dir + '/text.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

## 改行文字の削除

ファイルをテキストモードで開いて `read()` や `readline()` を呼び出すと、
`str` 型の文字列として読み込まれます。<br>

`str` 型の文字列の末尾にある改行文字が不要な場合は、
文字列に対して **`rstrip`** というメソッドを
`rstrip('\n')` のようにして呼び出すことにより削除することができます。
ここで、カッコ `()` の中にはそのテキストファイルで使われている改行文字を指定します。
一般的に、Windowsでは `\r\n`、macOSやLinuxでは `\n` です。
また、昔のmacOSで作られたテキストファイルは `\r` となっているものもあります。

```python
with open(pylearn_dir + '/novel.txt', 'r', encoding='utf-8') as f:
    while True:
        line = f.readline()
        if line == '':
            break
        print(line) # printが改行コードを付加する

print('------ 末尾の改行文字を削除すると以下のようになります-------')
with open(pylearn_dir + '/novel.txt', 'r', encoding='utf-8') as f:
    while True:
        line = f.readline()
        if line == '':
            break
        print(line.rstrip('\n'))
```
